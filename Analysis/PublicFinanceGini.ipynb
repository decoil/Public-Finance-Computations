```
giniDS <- read.csv("gini.csv") %>% na.omit()
pubSpend <- read.csv("PublicSpending.csv")
```

```
joinedPSgini <- pubSpend %>% 
    filter(TIME == 2019) %>% 
    select(LOCATION,Value) %>% 
    mutate(LOCATION = countrycode(LOCATION ,"iso3c", "country.name", nomatch=NULL)) %>%
    inner_join(giniDS, by = join_by("LOCATION" == "country")) %>%
    rename("publicspending" = "Value",
               "gini" = "value",
               "country" = "LOCATION")

joinedPSgini
```

```
options(repr.plot.width=15, repr.plot.height=10)

joinedPSgini %>% 
    ggplot( aes(publicspending, gini, color = country)) +
    geom_point()
```

```
shapiro.test(joinedPSgini$publicspending)
shapiro.test(joinedPSgini$gini)

cor.test(joinedPSgini$publicspending, joinedPSgini$gini)
cor.test(joinedPSgini$publicspending, joinedPSgini$gini, method="spearman", exact=FALSE)
```

Based on the results of the statistical tests, there may be a weak correlation between education expenditure and Gini. Using the Shapiro-Wilk normalitiy test we can see that the data for the education expenditure is normally distributed and the data for the Gini is not normally distributed. The normality of the data is an important distinction and may give us clues as to why the correlation is low. Moreover, using the Pearson correlation test, we find that the two variables are not correlated ($r = -0.051$), however, due to the high p-value ($p=0.78$) we can cannot conclude that the two variables are correlated. 

Likewise, using the Spearman correlation test, we find that there is a chance ($p=0.058$) that there is a weak negative correlation ($r = -0.339$) between the two variables. It is important to note that Spearman correlation test takes the rank of the variables, hence, it is more resistant to outliers. In our case, the Spearman method did provide for a weak correlation compared to the Pearson test. Going back to the normality test, we found that the data for the Gini was not normally distributed, which could mean that some of the countries are outliers and resultingly skew the correlation. Looking at the scatterplot and examining closer, we find that countries such as Costa Rica and Chile are notable outliers. This could suggest that the GDP of a country may also be an important factor and grouping by similar GDP could potentially provide for a stronger correlation between the Gini and education expenditure.

```
jps <- joinedPSgini %>% select(-country)

rmOutliers <- lapply(jps, function(x) {
  q1 <- quantile(x, .25)
  q2 <- quantile(x, .75)
  IQR <- q2 - q1
  replace(x, x < (q1 - 1.5*IQR) | x > (q2+1.5*IQR), NA)
})

rmFilt <- rmOutliers %>% 
    data.frame() %>% 
    cbind(country = joinedPSgini$country ) %>% 
    na.omit() 

#Removed the outliers Chile, Costa Rica, and Norways
```

```
shapiro.test(rmFilt$publicspending)
shapiro.test(rmFilt$gini)

cor.test(rmFilt$publicspending, rmFilt$gini)
cor.test(rmFilt$publicspending, rmFilt$gini, method="spearman", exact=FALSE)
```

After removing the statistical outliers, which included Costa Rica, Chile, and Norway, we find that the Gini is now normally distributed and there is statistically significant ($p=0.018$) moderate negative correlation ($r=-0.435$). Based on this information we can suggest a policy that will increase the amount of public spending on education. The negative correlation indicates that higher education expenditures will reduce the amount of income inequality (Gini). However, it is important to note that there is only a moderate correlation and it imply does not imply causation, hence, it is likely that other factors play a part in the income inequality and further investigations should be done.

Let $Y=400000$ be the income when healthy and when sick the income is represented as $Y-L=200000 \implies L=200000$ . The maximum insurance premium we are willing to pay is denoted by $I$ and assume that the insurance company will compensate for income losses when sick. We then compute $I$ as follows:
$$\begin{gathered} P_0Y + P_1(Y-L) = P_0(Y-I) + P_1(Y-I) \implies \\ \implies Y(P_0 + P_1) - P_1 L = Y-I(P_0+P_1) \implies \\ \implies Y - P_1L = Y-I \implies \\ \implies P_1L = I \implies I = (0.2) \cdot (200000) = 40000\end{gathered}$$

Hence, we would be willing to pay up to CZK 40000 for the insurance as this would at least bring us the same amount of utility as with no insurance.
